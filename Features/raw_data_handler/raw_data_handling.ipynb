#### Encoder dataset

```python
encoder_raw = pd.read_csv('../../datasets/raw_dataset/phishing_email.csv')
encoder_raw.shape

```

```python
SEED = 42

group_keys = ["text_combined"]
before = len(encoder_raw)
encoder_raw = encoder_raw.drop_duplicates(subset=group_keys,keep="first").reset_index(drop=True)
print(f"dedup removed: {before - len(encoder_raw)} | New shape: {encoder_raw.shape}")
```

```python
# Count class frequencies
class_counts = encoder_raw['label'].value_counts().sort_index()

# Plot
plt.figure(figsize=(6, 4))
plt.bar(class_counts.index.astype(str), class_counts.values, color='skyblue', edgecolor='black')
plt.title("Class Distribution", fontsize=14)
plt.xlabel("Class Label", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
X = encoder_raw.index.values
y = encoder_raw['label'].astype(int).values

# 80% train 20% val+test
idx_train, idx_tmp, y_train, y_tmp = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=SEED
)

# 10% val 10% test
idx_val, idx_test, y_val, y_test = train_test_split(
    idx_tmp, y_tmp, test_size=0.5, stratify=y_tmp, random_state=SEED
)

train_df = encoder_raw.loc[idx_train].reset_index(drop=True)
val_df   = encoder_raw.loc[idx_val].reset_index(drop=True)
test_df  = encoder_raw.loc[idx_test].reset_index(drop=True)
```

```python
print(train_df.dtypes)
print(val_df.dtypes)
print(test_df.dtypes)

print(train_df.shape)
print(val_df.shape)
print(test_df.shape)

print(len(set(val_df["text_combined"]) & set(test_df["text_combined"])))
print(len(set(val_df["text_combined"]) & set(train_df["text_combined"])))
print(len(set(test_df["text_combined"]) & set(train_df["text_combined"])))
```

```python
train_df.to_csv('../../datasets/encoder_dataset/raw/train_set.csv',index=False)
val_df.to_csv('../../datasets/encoder_dataset/raw/valid_set.csv',index=False)
test_df.to_csv('../../datasets/encoder_dataset/raw/test_set.csv',index=False)
```
